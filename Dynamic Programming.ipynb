***Question:*** what is the data in this structure saying? Relate this to the course
presentation of P

**Answer:**
In the course, P represnts the following probablistic function:
 P(s',r | s,a), which is the probablity of going into next state s' with reward
 r given being in state s and taking action a.

 The data structure env.env.P is showing us a dictionary with 16 values from
 0 to 15. each corresponding to being in state s in [0 to 15].

 Within each of the s values is another dictionary containing 4 tuples.
 Each tuple contains information corresponding to taking action 'a' while being in state 's'. Action 'a' can
 have a value between 0 to 3.

 Each tuple is of the following structure:

 [(probability of going into state s' given being in state s and taking action a,

 state s',

 reward r,

 boolean of whether done= true, meaning fell in Hole or acheive goal or done = false, meaning can keep on moving)]  


 In this case, the probability of going into state s' given being in state s and taking action a is always 1 because the "is_slippery" is set to False.

 From the openai gym wiki on github, the actions are as follows:

0	Move Left
1	Move Down
2	Move Right
3	Move Up

```

# Practical: revise the above AI solver to use a policy function in which you
# code the random action selections in the policy function. Test this.



# policy function
def init_policy(env):

  #initiate policy to code the random action selections within it
  policy = np.zeros((env.observation_space.n))

  for i in range(env.observation_space.n):
    policy[i]=int(env.action_space.sample())
  return policy


env.reset()
exitCommand=False

observation=""
prev_observation="" # to hold information about the current state that the agent is in
policy=init_policy(env) #initiate the policy
while not(exitCommand):
  env.render()

  if observation =="":
    prev_observation=0
  else:
    prev_observation=observation

  #get action from initiated policy
  action = int(policy[prev_observation])

  (observation, reward, compute, probability) = env.step(action)
  print("--> The result of taking action",action,"is:")
  print("     S=",observation)
  print("     R=",reward)
  print("     p=",probability)


  #get info about exitCommand (whether fell in hole or reached goal)
  exitCommand = env.env.P[prev_observation][action][0][3]
  env.render()
```

```
# Practical: Code the C-4 Policy Evaluation (Prediction) algorithm. You may use
# either the inplace or ping-pong buffer (as described in the lecture). Now
# randomly initialize your policy function, and compute its value function.
# Report your results: policy and value function. Ensure your prediction
# algo reports how many iterations it took.


# initialize policy function
def init_random_policy():
  # initiate policy as an equiprobable randomly initialized policy
  policy=np.ones([env.observation_space.n, env.action_space.n]) / env.action_space.n
  return policy

# initialize state value function
def init_state_value():
  # initialize state value function as 0 for all states
  V=np.zeros(env.observation_space.n)
  return V


#theta is threshold for accuracy
#gamma is discounting factor, this is epoisodic so can have gamma=1
def policy_eval(policy, env, theta, gamma):

  #initialize V(s) as zeroes
  V=init_state_value()
  v=0
  iterations=0
  while True:
    iterations+=1
    delta =0
    for s in range(env.observation_space.n):
      v=V[s]
      V[s]=0 # reset to 0 because will recalculate
      for a in range(env.action_space.n):
        for p, s_prime, r, exitCommand in env.env.P[s][a]:
          V[s] += (policy[s][a] * p * (r + (gamma*V[s_prime])))
      delta = max(delta, abs(v-V[s]))

    if delta < theta:
      break

  return V, iterations


policy = init_random_policy()
V=policy_eval(policy,env,0.0001,1)

print("The Value Function is: ")
print()
print(V[0])
print("The number of iterations is: ", V[1])
```
